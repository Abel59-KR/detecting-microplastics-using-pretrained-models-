```python
import numpy as np # linear algebra
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
import cv2
for dirname, _, filenames in os.walk('C:\\Users\\User\\Desktop\\comp 403\\data'):
    print(os.path.join(dirname))
```

```python

def get_data(path):
    images = []
    filenames = []
    df = None

    for root, dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.join(root, file)

            if file.endswith('.jpg'):
                img = Image.open(file_path)
                images.append(img)
                filenames.append(file)
            elif file.endswith('.csv'):
                df = pd.read_csv(file_path)

    return images, filenames, df
```

```python
import math
def estimate_noise(I):
    H, W = I.shape
    M = [[1, -2, 1], [-2, 4, -2], [1, -2, 1]]
    sigma = np.sum(np.sum(np.absolute(cv2.filter2D(I, -1, np.array(M)))))
    sigma = sigma * math.sqrt(0.5 * math.pi) / (6 * (W - 2) * (H - 2))
    return sigma

def calculate_average_noise_variance(image_folder):
    total_noise = 0
    num_images = 0

    for filename in os.listdir(image_folder):
        if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
            img_gray = cv2.imread(os.path.join(image_folder, filename), cv2.IMREAD_GRAYSCALE)
            noise = estimate_noise(img_gray)
            total_noise += noise
            num_images += 1

    if num_images > 0:
        average_noise = total_noise / num_images
        print(f"Average noise variance across {num_images} images in folder '{image_folder}': {average_noise:.2f}")
    else:
        print(f"No valid images found in the folder '{image_folder}'.")

# Specify the path to your image folders
valid2_folder_path = "C:\\Users\\User\\Desktop\\comp 403\\data2\\valid2"
train2_folder_path = "C:\\Users\\User\\Desktop\\comp 403\\data2\\train2"
valid_folder_path = "C:\\Users\\User\\Desktop\\comp 403\\data\\valid"
train_folder_path = "C:\\Users\\User\\Desktop\\comp 403\\data\\train"
calculate_average_noise_variance(valid2_folder_path)
calculate_average_noise_variance(train2_folder_path)
calculate_average_noise_variance(valid_folder_path)
calculate_average_noise_variance(train_folder_path)
```
